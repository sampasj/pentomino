```python
def solve() :
    global masume
    global use_piece
    global m
    for x in range(75) :
        if masume[x] == 0 :
            for s in sugata :
                if possible(x, s) :
                    use_piece.append(s[5])
                    for i in range(5) :
                        masume[x + s[i]] = s[5]
                    solve()
                    for i in range(5) :
                        masume[x + s[i]] = 0
                    use_piece.pop()
            return
    m += 1
    print("m=",m)
    seikei(masume)
```
